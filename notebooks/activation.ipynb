```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sig(x: float) -> float:
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sig(x)
```

```python
def sig_derivative(x):
    s = sig(x)
    return s * (1 - s)

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)
```

```python
plt.plot(x, sig(x), label="σ(x)")
plt.plot(x, sig_derivative(x), label="σ'(x)")
plt.title("Sigmoid and its derivative")
plt.xlabel("x")
plt.legend()
plt.grid(True)
plt.show()
plt.plot(x, relu(x), label="ReLU(x)")
plt.plot(x, relu_derivative(x), label="ReLU'(x)")
plt.title("ReLU and its derivative")
plt.xlabel("x")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

def softplus(x):
    return np.log(1 + np.exp(x))

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2/np.pi) * (x + 0.044715 * x**3)))

x = np.linspace(-5, 5, 100)

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].plot(x, sig(x))
axes[0, 0].set_title("Sigmoid: Saturating, smooth")
axes[0, 0].grid(True)
axes[0, 1].plot(x, tanh(x))
axes[0, 1].set_title("Tanh: Saturating, zero-centered")
axes[0, 1].grid(True)
axes[1, 0].plot(x, relu(x), label="ReLU")
axes[1, 0].plot(x, softplus(x), label="Softplus")
axes[1, 0].set_title("Softplus: Smooth ReLU")
axes[1, 0].legend()
axes[1, 0].grid(True)
axes[1, 1].plot(x, relu(x), label="ReLU")
axes[1, 1].plot(x, gelu(x), label="GELU")
axes[1, 1].set_title("GELU: Used in transformers")
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def softplus_derivative(x):
    return 1 / (1 + np.exp(-x))  

def gelu_derivative(x):
    
    h = 0.0001
    return (gelu(x + h) - gelu(x - h)) / (2 * h)

x = np.linspace(-5, 5, 100)

plt.figure(figsize=(12, 5))
plt.plot(x, sig_derivative(x), label="Sigmoid' (max 0.25)")
plt.plot(x, tanh_derivative(x), label="Tanh' (max 1.0)")
plt.plot(x, softplus_derivative(x), label="Softplus' (0 to 1)")
plt.plot(x, relu_derivative(x), label="ReLU' (0 or 1)")
plt.plot(x, gelu_derivative(x), label="GELU'", linestyle='--')
plt.title("Derivative comparison")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.legend()
plt.grid(True)
plt.show()
```
